```python
import pandas as pd
import numpy as np

data = pd.read_csv('../../data/processed/master_dataset.csv')
data.head()
```

```python
preferred_aggregations = {
    'HVGDP': '99',
    'UVG5': '99', # Gross value added at current prices - services
    'UVGM': '99', # Gross value added at current prices - manufacturing industry
    'UVG1': '99', # Gross Value Added at current prices - agriculture, forestry and fishery products,
    'UXGS': '99', # Exports of goods and services at current prices (National accounts)
    'UMGS': '99', # Imports of goods and services at current prices (National accounts),
    'PIGT': '99', # Price deflator gross fixed capital formation: total economy,
    'NETD': '0', # Employment, persons: all domestic industries (National accounts),
    'UDGGL': '99'
}
countries_of_interest = ['North Macedonia', 'Serbia', 'Albania', 'Greece', 'Bulgaria', 'Montenegro', 'Croatia', 'Slovenia', 'European Union']
mask = pd.Series(False, index=data.index)

for var, agg in preferred_aggregations.items():
    pattern = rf'\.{agg}\..*\.{var}$'
    mask |= data['SERIES'].str.contains(pattern, regex=True)

filtered_data = data[mask]
filtered_data = filtered_data[filtered_data['COUNTRY'].isin(countries_of_interest)]


filtered_data
```

```python
data = filtered_data[['COUNTRY', 'CODE', 'TITLE'] + [str(x) for x in range(2000, 2024)]]
data
```

```python
data = data.melt(id_vars=['CODE', 'COUNTRY', 'TITLE'], var_name='DATE', value_name='VALUE')
data
```

```python
eu_gdp = data[(data['COUNTRY']=='European Union') & (data['CODE']=='HVGDP')][['DATE','VALUE']]
eu_gdp.columns = ['DATE', 'EU_HVGDP']
eu_gdp
```

```python
data = data.pivot_table(
    index=['COUNTRY', 'DATE'],
    columns=['CODE'],
    values=['VALUE']
).reset_index()
data
```

```python
wgi_df = pd.read_excel('../../data/original/wgidataset.xlsx')

balkan_codes = ['ALB', 'BGR', 'HRV', 'MNE', 'MKD', 'SRB', 'SVN', 'GRC']

wgi_balkan_df = wgi_df[
    (wgi_df['code'].isin(balkan_codes)) &
    (wgi_df['year'] >= 2000) & (wgi_df['year'] <= 2024)
].copy()

wgi_balkan_df = wgi_balkan_df.pivot_table(
    index=['countryname', 'year'],
    columns='indicator',
    values=['estimate'],
    aggfunc='first'
)

wgi_balkan_df.columns = ['_'.join(col).strip() for col in wgi_balkan_df.columns.values]
wgi_balkan_df = wgi_balkan_df.reset_index()

wgi_balkan_df
```

Meaning of variables:
- CC: Control of Corruption
- GE: Government Effectiveness
- RL: Rule of Law
- RQ: Regulatory Quality
- PV: Political Stability and Absence of Violence/Terrorism
- VA: Voice and Accountability

```python
wgi_balkan_df['year'] = wgi_balkan_df['year'].astype(str)
final_df = final_df.merge(wgi_balkan_df, left_on=['COUNTRY', 'DATE'], right_on=['countryname', 'year'], how='inner')
```

```python
final_features_df = final_df.copy()

final_features_df["Total_Trade"] = final_features_df["UMGS"] + final_features_df["UXGS"]
final_features_df["Trade_Openness"] = final_features_df["Total_Trade"] / final_features_df["HVGDP"]
final_features_df["Net_Exports"] = final_features_df["UXGS"] - final_features_df["UMGS"]
final_features_df["Net_Exports_GDP"] = final_features_df["Net_Exports"] / final_features_df["HVGDP"]

final_features_df['Prod_Agriculture'] = (final_features_df['UVG1']) / final_features_df['Employees_Agriculture']
final_features_df['Prod_Industry'] = (final_features_df['UVGM']) / final_features_df['Employees_Industry']
final_features_df['Prod_Services'] = (final_features_df['UVG5']) / final_features_df['Employees_Services']

final_features_df['Share_Agriculture'] = final_features_df['Employees_Agriculture'] / final_features_df['Employees_Total']
final_features_df['Share_Industry'] = final_features_df['Employees_Industry'] / final_features_df['Employees_Total']
final_features_df['Share_Services'] = final_features_df['Employees_Services'] / final_features_df['Employees_Total']

g = final_features_df.groupby('COUNTRY')
final_features_df['dShare_Agr'] = g['Share_Agriculture'].diff()
final_features_df['dShare_Ind'] = g['Share_Industry'].diff()
final_features_df['dShare_Serv'] = g['Share_Services'].diff()

final_features_df['Prod_Agr_lag'] = g['Prod_Agriculture'].shift()
final_features_df['Prod_Ind_lag'] = g['Prod_Industry'].shift()
final_features_df['Prod_Serv_lag'] = g['Prod_Services'].shift()

final_features_df['SC'] = (
    final_features_df['dShare_Agr'] * final_features_df['Prod_Agr_lag'] +
    final_features_df['dShare_Ind'] * final_features_df['Prod_Ind_lag'] +
    final_features_df['dShare_Serv'] * final_features_df['Prod_Serv_lag']
)

final_features_df['dProd_Agr'] = g['Prod_Agriculture'].diff()
final_features_df['dProd_Ind'] = g['Prod_Industry'].diff()
final_features_df['dProd_Serv'] = g['Prod_Services'].diff()

final_features_df['Within_SP'] = (
    g['Share_Agriculture'].shift() * final_features_df['dProd_Agr'] +
    g['Share_Industry'].shift() * final_features_df['dProd_Ind'] +
    g['Share_Services'].shift() * final_features_df['dProd_Serv']
)

final_features_df['Total_Prod'] = (
    final_features_df['Share_Agriculture'] * final_features_df['Prod_Agriculture'] +
    final_features_df['Share_Industry'] * final_features_df['Prod_Industry'] +
    final_features_df['Share_Services'] * final_features_df['Prod_Services']
)
final_features_df['dTotal_Prod'] = g['Total_Prod'].diff()

final_features_df['GDP_growth'] = g['HVGDP'].pct_change()

final_features_df['EU_Distance'] = (final_features_df['HVGDP'] / final_features_df['EU_HVGDP']) - 1

final_features_df['log_GDP'] = np.log(final_features_df['HVGDP'])
final_features_df['log_EU_GDP'] = np.log(final_features_df['EU_HVGDP'])
final_features_df['log_GDP_gap'] = final_features_df['log_EU_GDP'] - final_features_df['log_GDP']

final_features_df['Employment_growth'] = g['NETD'].pct_change()

final_features_df = final_features_df.dropna()
final_features_df
```

```python
print(f"Total observations: {len(final_features_df)}", end="\n\n")
print("Observations by country:\n", final_features_df.groupby('COUNTRY').size().sort_values(ascending=False))

print("\nSC + Within_SP ~= dTotal_Prod (McMillan-Rodrik identity):")
final_features_df['Decomposition_Check'] = final_features_df['SC'] + final_features_df['Within_SP']
final_features_df['Decomp_Error'] = abs(final_features_df['dTotal_Prod'] - final_features_df['Decomposition_Check'])

print(f"Mean absolute decomposition error: {final_features_df['Decomp_Error'].mean():.2f}")
print(f"Max absolute decomposition error: {final_features_df['Decomp_Error'].max():.2f}")
print(f"Correlation (SC+Within vs dTotal_Prod): {final_features_df['Decomposition_Check'].corr(final_features_df['dTotal_Prod']):.4f}")

print("\nSample check (first 5 rows):")
check_cols = ['COUNTRY', 'DATE', 'SC', 'Within_SP', 'Decomposition_Check', 'dTotal_Prod', 'Decomp_Error']
print(final_features_df[check_cols].head(), end='\n\n')

key_vars = ['EU_Distance', 'SC', 'Within_SP', 'Trade_Openness', 'GDP_growth', 'PIGT', 'NETD']
print(final_features_df[key_vars].describe())
```
